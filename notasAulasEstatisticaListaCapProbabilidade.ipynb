2. Considere um jogo com bolas numeradas de 1 a $4$, $\Omega  = \{1,\; 2,\; 3,\;4,\,5,\;6\}$ em uma esfera giratória. No processo de escolha temos as bolas tem mesma chance de seres escolhidas. Sabendo disso calcule:
   1. as probabilidades de cada bola ser escolhida.
   2. a probabilidade do evento $A = \{x|x \text{ a bola tem número par} \}$.
   3. a probabilidade do evento $B = \{x|x \text{ a bola tem número impar} \}$.
   4. a probabilidade do evento $C = \{x|x \text{ a bola tem número menor que } 3 \}$.
   5. as probabilidades
        1. $P(B)$
        2. $P(A \cup B)$
        3. $P(A \cap B)$
        3. $P(A \cup B \cup C)$

```python
# A. as probabilidades de cada bola ser escolhida são:
display(Markdown(f'$P({1})=P({2})=P({3})=P({4})=P({5})=P({6})=\\displaystyle \\frac{1}{6}$'))
# 2. a probabilidade do evento $A = \{x|x \text{ a bola tem número par} \}$.
omega_dado = list(range(1,7))
evento_A = list(range(2,7,2))
display(Markdown(f'$P(A) = {len(evento_A)/len(omega_dado)}$'))
# 3. a probabilidade do evento $B = \{x|x \text{ a bola tem número impar} \}$.
evento_B = list(range(1,7,2))
display(Markdown(f'$P(B) = {len(evento_B)/len(omega_dado)}$'))

# 4. a probabilidade do evento $C = \{x|x \text{ a bola tem número menor que } 3 \}$.
evento_C = list(range(1,3))
display(Markdown(f'$P(C) = {len(evento_C)/len(omega_dado)}$'))
# 5. as probabilidades
#        1. $P(B)$
evento_B = list(range(1,7,2))
display(Markdown(f'$P(B) = {len(evento_B)/len(omega_dado)}$'))
#        2. $P(A \cup B)$
evento_B = list(range(1,7,2))
display(Markdown(f'$P(A\cup B) = {len(evento_A)/len(omega_dado)+len(evento_B)/len(omega_dado)}$'))
#        3. $P(A \cap B)$
display(Markdown(f'$P(A\cap B) = {1-(len(evento_A)/len(omega_dado)+len(evento_B)/len(omega_dado))}$'))
#        3. $P(A \cup B \cup C)$
#
display(Markdown(f'$P(A\cap B\cup C) = {(len(omega_dado)/len(omega_dado))}$'))
```
